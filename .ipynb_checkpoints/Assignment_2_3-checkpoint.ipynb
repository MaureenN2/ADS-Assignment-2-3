```python
#import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
employee_attrition = pd.read_csv(r"C:\Users\User\Downloads\Assignment\ADS-Assignment-2-3\ADS-Assignment-2-3\WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
#observe dataset
employee_attrition.head()
```

```python
#observe dataset
employee_attrition.shape
```

```python
#Statistical description
employee_attrition.describe(include = 'all')
```

```python
employee_attrition['JobRole'].value_counts()
```

```python
#Q2. Aggregating dataset by groupby and find the mean
tk2 =employee_attrition.groupby (['JobRole', 'Attrition'])['DistanceFromHome'].mean().reset_index()
tk2
```

```python
#Using group by Agggregation to plot
plt.figure(figsize =(10,7))
sns.barplot(x="JobRole", y="DistanceFromHome", hue = 'Attrition',  data=tk2)
plt.xticks(rotation = 90)
plt.title("Job Role Against Distance From Home per Attrition")
plt.xlabel("JobRole")
plt.ylabel("Distance From Home")
```

```python
import numpy as np
```

```python
#using estimatior
sns.barplot(x = 'JobRole' , y = 'DistanceFromHome', data = employee_attrition, hue = 'Attrition', estimator = np.mean)
plt.xticks(rotation = 90)
plt.title("Job Role Against Distance From Home per Attrition")
plt.xlabel("JobRole")
plt.ylabel("Distance From Home")
```

```python
#quick analysis of value count
employee_attrition['Education'].value_counts()
```

```python
#aggregating and represeting the dataset on table format
tk3 =employee_attrition.groupby  ([ 'Education', 'Attrition'])['MonthlyIncome'].mean().reset_index()
tk3
```

```python
#plotting using the aggregated data
sns.barplot(x="Education", y="MonthlyIncome", hue="Attrition", data=tk3)
plt.title("Education Against MonthlyIncome per Attrition")
plt.xlabel("Education")
plt.ylabel("Monthly Income")
```
